```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 録音したWAVファイルのパスを指定
audio_path = 'instrument_sample.wav'

# 音声の読み込み (y: 音声データ, sr: サンプリングレート)
y, sr = librosa.load(audio_path)

# グラフを描画するウィンドウの準備
plt.figure(figsize=(10, 8))

# ==========================================
# 1. 時間波形のグラフ (ADSRの分析用)
# ==========================================
plt.subplot(2, 1, 1) # 2行1列の上のグラフ
librosa.display.waveshow(y, sr=sr, alpha=0.8)
plt.title('Waveform (ADSR Analysis - Check Attack and Release times)')
plt.ylabel('Amplitude (Volume)')
plt.xlabel('Time (Seconds)')

# ==========================================
# 2. スペクトログラム (倍音構造の分析用)
# ==========================================
plt.subplot(2, 1, 2) # 2行1列の下のグラフ
# 音声の周波数を分析 (STFT変換)
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
# グラフに描画 (y軸を対数スケールにして人間の耳の感覚に合わせる)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram (Overtone Analysis - Check high frequencies)')
plt.xlabel('Time (Seconds)')
plt.ylabel('Frequency (Hz)')

plt.tight_layout()z
```
